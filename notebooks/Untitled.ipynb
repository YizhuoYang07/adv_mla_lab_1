```python
import pandas as pd
import numpy as np
from joblib import dump
```

```python
X_train = pd.read_csv('data/processed/X_train.csv')
X_val   = pd.read_csv('data/processed/X_val.csv'  )
X_test  = pd.read_csv('data/processed/X_test.csv' )
y_train = pd.read_csv('data/processed/y_train.csv')
y_val   = pd.read_csv('data/processed/y_val.csv'  )
y_test  = pd.read_csv('data/processed/y_test.csv' )
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf1 = RandomForestClassifier(random_state=8)
```

```python
rf1.fit(X_train, y_train)
```

```python
from joblib import dump

dump(rf1,  'models/rf1.joblib')
```

```python
y_train_probas = rf1.predict_proba(X_train)
y_val_probas = rf1.predict_proba(X_val)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_train_pred = y_train_probas[:, 1]

roc_score_train = roc_auc_score(y_train, y_train_pred)
print(f"Training ROC AUC score: {roc_score_train:.4f}")
```

```python
y_val_pred = y_val_probas[:, 1]

roc_score_val = roc_auc_score(y_val, y_val_pred)
print(f"Validation ROC AUC score: {roc_score_val:.4f}")
```

```python
rf2 = RandomForestClassifier(random_state=8, max_depth=6, min_samples_leaf=50)
rf2.fit(X_train, y_train)
```

```python
print(roc_auc_score(y_train, rf2.predict_proba(X_train)[:, 1]))
print(roc_auc_score(y_val, rf2.predict_proba(X_val)[:, 1]))
```

```python
dump(rf2,  'models/rf2.joblib')
```
